```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID column 'NAME'.
application_df = application_df.drop(axis=1, columns="NAME")
```

```python
# Drop the non-beneficial ID column 'EIN'.
application_df = application_df.drop(axis=1, columns="EIN")
```

```python
application_df["ASK_AMT"].value_counts()
```

```python
#analyse split of is-successful - these will form our y values
import matplotlib.pyplot as plt
y = k["IS_SUCCESSFUL"].value_counts()
y.plot(kind="bar", figsize=(6,4), color = "b", alpha=1, width=0.85, rot=90)
plt.title("Loaned 5K Success Rate")
plt.xlabel("Result")
plt.ylabel("Count of Result")
plt.show()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

application_types_to_replace = ["T8","T7","T10","T9","T13","T12","T2","T25","T14","T15","T29","T17"]

["T9","T13","T12","T2","T25","T14","T15","T29","T17"]

#T8         737
#T7         725
#T10        528
#Other      276


# Replace in dataframe
for app in application_types_to_replace:
    
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_count = application_df["CLASSIFICATION"].value_counts()

class_count.loc[class_count > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = class_count.loc[class_count < 1883].index

print(classifications_to_replace)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df.nunique().sum()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
df = pd.get_dummies(application_df)
```

```python
# Split our preprocessed data into our features and target arrays

y = df["IS_SUCCESSFUL"]

X = df.drop(axis=1, columns="IS_SUCCESSFUL")

# Split the preprocessed data into a training and testing dataset

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X.columns.value_counts()
```

# Optimisation 1 - Increase binned values for applicationt type

```python
# Initialise the Sequential class
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=5, activation="relu",input_dim=40))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=5, activation="relu",input_dim=40))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy", "AUC"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
history_df = pd.DataFrame(fit_model.history)

history_df.head()
```

```python
import matplotlib.pyplot as plt

# Plot the loss
history_df.plot(y="loss")
plt.show()
```

```python
# Plot the accuracy
history_df.plot(y="accuracy")
plt.show()
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy, model_auc = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
